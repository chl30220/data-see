### 데이터 만들기

```python
import matplotlib.pyplot as plt
```

```python
a=[1,2,3,4,-5,5,10,-7]
b=['A','B','C','D','E','G','H','I']
```

```python
plt.bar(b,a, color='red')
```

```python
plt.barh(b,a, color='green')
```

```python
plt.plot(b,a, 'rx-')
plt.title('subject')
plt.xlabel('school')
plt.ylabel('score')
plt.legend(['jumsu'])
```
